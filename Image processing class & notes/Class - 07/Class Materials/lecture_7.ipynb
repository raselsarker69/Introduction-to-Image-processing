```python
# Sobel filtering

import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = cv.imread('/content/drive/MyDrive/image_processing_innovative_skills/images/sudoku-1.jpg', cv.IMREAD_GRAYSCALE)

sobel_x = cv.Sobel(img, -1, 1, 0, 3)
sobel_y = cv.Sobel(img, -1, 0, 1, 3)

plt.figure(figsize=(15,15))

plt.subplot(1, 3, 1)
plt.imshow(img, cmap='gray')
plt.title('Original')
plt.axis('off')


plt.subplot(1, 3, 2)
plt.imshow(sobel_x, cmap='gray')
plt.title('vertical sobel filter')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(sobel_y, cmap='gray')
plt.title('horizontal sobel filter')
plt.axis('off')

plt.show()
```

```python
sobel = cv.Sobel(img, -1, 1, 1, 3)
plt.imshow(sobel, cmap='gray')
plt.title('Sobel')
plt.axis('off')
```

```python
# sobel vs laplacian

import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = cv.imread('/content/drive/MyDrive/image_processing_innovative_skills/images/sudoku-1.jpg', cv.IMREAD_GRAYSCALE)

sobel_x = cv.Sobel(img, -1, 1, 0, 3)
sobel_y = cv.Sobel(img, -1, 0, 1, 3)
laplacian = cv.Laplacian(img, -1)


plt.figure(figsize=(10,8))

plt.subplot(2, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original')
plt.axis('off')


plt.subplot(2, 2, 2)
plt.imshow(sobel_x, cmap='gray')
plt.title('vertical sobel filter')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(sobel_y, cmap='gray')
plt.title('horizontal sobel filter')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(laplacian, cmap='gray')
plt.title('laplacian filter')
plt.axis('off')

plt.show()
```

```python
# Canny Edge Detection

import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Load the image
img = cv.imread('/content/drive/MyDrive/image_processing_innovative_skills/images/sudoku-1.jpg', cv.IMREAD_GRAYSCALE)

# Apply Canny edge detection
lower_threshold = 100
upper_threshold = 200
edges = cv.Canny(img, lower_threshold, upper_threshold)


plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(edges, cmap='gray')
plt.title('Canny Edge Detection')
plt.axis('off')

plt.tight_layout()
plt.show()
```
