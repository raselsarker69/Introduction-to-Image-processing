```python
image = cv2.imread('flower.jpg')

blue, green, red = cv2.split(image)

rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(6, 10))

plt.subplot(2, 2, 1)
plt.imshow(rgb)
plt.title('RGB Image')

plt.subplot(2, 2, 2)
plt.imshow(blue, cmap='gray')
plt.title('blue channel')

plt.subplot(2, 2, 3)
plt.imshow(green, cmap='gray')
plt.title('green channel')

plt.subplot(2, 2, 4)
plt.imshow(red, cmap='gray')
plt.title('red channel')
```

Explain the reason of color difference when an image is loaded with openCV but plotted using matplotlib.

The color difference when loading an image with OpenCV and displaying it with Matplotlib occurs because,,,,,,

OpenCV: loads images in BGR format, meaning that the first channel is blue, the second is green, and the third is red.
Matplotlib: on the other hand, expects images in RGB format, where the first channel is red, the second is green, and the third is blue.
Because the library controls two different color channels by default, that's the main reason for the difference in color.

```python
import matplotlib.pyplot as plt
import cv2

# Paths to the images
image_path1 = 'flower.jpg'
image_path2 = 'flower.jpg'
image_path3 = 'flower.jpg'

# Load images using OpenCV
image1 = cv2.imread(image_path1)
image2 = cv2.imread(image_path2)
image3 = cv2.imread(image_path3)

# Convert from BGR to RGB for correct display
image1_rgb = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
image2_rgb = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
image3_rgb = cv2.cvtColor(image3, cv2.COLOR_BGR2RGB)

# Create a figure
fig, axes = plt.subplots(3, 1, figsize=(5, 15))  # 3 rows, 1 column

# Plot image in a different subplot
axes[0].imshow(image1_rgb)
axes[0].axis('off')  

axes[1].imshow(image2_rgb)
axes[1].axis('off')

axes[2].imshow(image3_rgb)
axes[2].axis('off')

# Show the figure
plt.show()
```
