# Image Loding and showing using Matplotlib

```python
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
# image loading.
image_path = 'flower.jpg'

# Load image
image = plt.imread(image_path)


plt.imshow(image)  # Display image
#plt.axis('off')  # hide axes
plt.show()
```

# Imag loard korbo Opencv diya and show korabo Matplotlib diya.

```python
import cv2
image_path = 'flower.jpg'

# Load image
image = cv2.imread(image_path)


plt.imshow(image)  # Display image
#plt.axis('off')  # hide axes
plt.show()
```

Why does the image look different?

OpenCV uses BGR format, so when we read an image using cv2.imread() interrupts BGR format by default, opencv image in BGR and we are floating RGB image so we need to convert the BGR to RGB format.

# Color space conversion BGR TO RGB

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
cv2.imshow('RGB Image',image_rgb )
```

```python
image = cv2.imread('flower.jpg')

blue, green, red = cv2.split(image)

rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(6, 10))

plt.subplot(2, 2, 1)
plt.imshow(rgb)
plt.title('RGB Image')

plt.subplot(2, 2, 2)
plt.imshow(blue, cmap='gray')
plt.title('blue channel')

plt.subplot(2, 2, 3)
plt.imshow(green, cmap='gray')
plt.title('green channel')

plt.subplot(2, 2, 4)
plt.imshow(red, cmap='gray')
plt.title('red channel')
```

# PIL(Pillow)

```python
from PIL import Image
import matplotlib.pyplot as plt 

image_path = 'flower.jpg'

# Load image
image = Image.open(image_path)


plt.imshow(image)  # Display image
#plt.axis('off')  # hide axes
plt.show()
```
