# 1. Binary Thresholding ( THRESH_BINARY )

```python
import cv2
import matplotlib.pyplot as plt

# Read the image as grayscale
img_path = cv2.imread('berry-1.jpg', cv2.IMREAD_GRAYSCALE)  # 0 indicates grayscale mode


# Apply the threshold function
th, res = cv2.threshold(img_path, 120, 255, cv2.THRESH_BINARY)

# Use matplotlib to display the image
plt.imshow(res, cmap='gray')
plt.title('Threshold Image')
plt.axis('off')  # Turn off axis labels
plt.show()
```


When you use cv2.imread('berry-1.jpg', 0) in OpenCV, the image is read in grayscale mode, meaning each pixel will have a single intensity value representing different shades of gray. In this mode, the pixel values typically range from 0 to 255, where:

0 represents black (the lowest intensity).
                    
255 represents white (the highest intensity).
                      
Values in between represent shades of gray.

```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

img = cv.imread('berry-1.jpg', cv.IMREAD_GRAYSCALE)

# cv.threshold(gray_img, threshold_val, max_val, threshold_type)

# ret,thresh1 = cv.threshold(img, 190,255, cv.THRESH_BINARY)
ret,thresh1 = cv.threshold(img, 130,255, cv.THRESH_BINARY)
ret,thresh2 = cv.threshold(img, 120,255, cv.THRESH_BINARY_INV)

ret,thresh3 = cv.threshold(img, 127,255, cv.THRESH_TRUNC) # >threshold = threshold; <= unchanged
ret,thresh4 = cv.threshold(img, 127,255, cv.THRESH_TOZERO) # >threshold unchanged; <= 0(black);
ret,thresh5 = cv.threshold(img, 127,255, cv.THRESH_TOZERO_INV) # >threshold = 0(black); <= unchanged

titles = ['Original Image','BINARY','BINARY_INV','TRUNC','TOZERO','TOZERO_INV']
images = [img, thresh1, thresh2, thresh3, thresh4, thresh5]

for i in range(6):
    plt.subplot(2,3,i+1)
    plt.imshow(images[i], 'gray',vmin=0,vmax=255)
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])

plt.show()
```
