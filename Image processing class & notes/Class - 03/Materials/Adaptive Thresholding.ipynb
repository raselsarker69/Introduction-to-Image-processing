### Adaptive Thresholding.

### 1.what is adaptive thresholding?
- A thresholding technique that finds threshold value based on a local region of pixels.

### 1. Why Do We Need Adaptive Thresholding?
- Uneven Lighting.
- Varying Contrast.
- Obtain outline of image.

### Compare between global threshold and adaptive threshold:
- Global Thresholding: Good for images with even lighting and high contrast.
- Adaptive Thresholding: Necessary for images with uneven lighting, shadows, or regions of low contrast. It ensures that the threshold is computed dynamically based on local intensity variations, providing better separation of objects from the background.

### 3.how does adaptive thresholding work?
- Global thresholding can indeed be simple and effective in certain cases, such as document scanning. However, the problem arises when there is uneven lighting in the image. Global thresholding uses a single threshold value that is applied uniformly across the entire image. As a result, when there is a variation in lighting intensity, some areas may become too bright or too dark, making accurate image processing difficult.

- Since document scanning often involves varying light intensities across different areas, global thresholding doesn't always yield the best results. To address the issue of uneven lighting, adaptive thresholding can be much more effective. It sets the threshold based on the intensity of light in different parts of the image, leading to more accurate outcomes for document scanning.

### Why adaptive thresholding works better:

- Handling uneven lighting: Adaptive thresholding divides the image into smaller blocks and sets a different threshold for each block based on the local lighting conditions. This allows for more accurate image processing, even with varying lighting intensities.
- Preserving small details: Adaptive thresholding is also helpful in preserving small details like text or patterns, ensuring that the finer elements of the image are captured correctly.

```python
import cv2 as cv
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 

img_path = cv.imread('berry-1.jpg', cv.IMREAD_GRAYSCALE)

ret, th1 = cv.threshold(img_path, 130, 255, cv.THRESH_BINARY)
th2 = cv.adaptiveThreshold(img_path, 255, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, 11, 2) 
th3 = cv.adaptiveThreshold(img_path, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY, 11, 2) 


titles = (['Original Image', 'Global img (v=60)', 'Adaptive threshold', 'Adaptive gushassian'])
images = (img_path, th1, th2, th3)


for i in range(4): 
    plt.subplot(2,2,i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```
