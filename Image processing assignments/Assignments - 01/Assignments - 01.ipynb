# HW: 02  Image loading and showing using at least one of these ( openCV, Skimage)

```python
import cv2
from matplotlib import pyplot as plt

# Load the image
image = cv2.imread('flower.jpg')

# Convert the image from BGR to RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display the image
plt.imshow(image_rgb)
plt.axis('off')  # Hide axes
plt.show()
```

# HW-3: explain the reason of color difference when an image is loaded with openCV but plotted using matplotlib.

The color difference when loading an image with OpenCV and displaying it with Matplotlib occurs because,,,,,,

OpenCV: loads images in BGR format, meaning that the first channel is blue, the second is green, and the third is red. 

Matplotlib: on the other hand, expects images in RGB format, where the first channel is red, the second is green, and the third is blue. Because the library controls two different color channels by default, that's the main reason for the difference in color.

# Bonus-2: Plot three images in a single column using matplotlib.

```python
image_path1 = 'flower.jpg'
image_path2 = 'flower.jpg'
image_path3 = 'flower.jpg'


image1 = plt.imread(image_path1)
image2 = plt.imread(image_path2)
image3 = plt.imread(image_path3)

fig, axes = plt.subplots(3, 1, figsize=(5, 15))  # 3 rows, 1 column

# List of images
images = [image1, image2, image3]

# Plot each image in subplot
for ax, img in zip(axes, images):
    ax.imshow(img)
    ax.axis('off')  

plt.show()
```

```python
image_path = 'flower.jpg'
image = Image.open(image_path)

image_np = np.array(image)

R, G, B = image_np[:, :, 0], image_np[:, :, 1], image_np[:, :, 2]   # Split into R, G, B channels


plt.figure(figsize=(20, 5))


plt.subplot(1, 4, 1)    # 1 row, 4 columns, position 1
plt.imshow(image_np)
plt.title("Original Image")
plt.axis('off')


plt.subplot(1, 4, 2)    # position 2
plt.imshow(R, cmap='Reds')
plt.title("Red Channel")
plt.axis('off')


plt.subplot(1, 4, 3)    #position 3
plt.imshow(G, cmap='Greens')
plt.title("Green Channel")
plt.axis('off')


plt.subplot(1, 4, 4)    #position 4
plt.imshow(B, cmap='Blues')
plt.title("Blue Channel")
plt.axis('off')

plt.show()
```
