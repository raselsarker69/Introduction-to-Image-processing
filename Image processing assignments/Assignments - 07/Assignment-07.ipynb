### HW: 1. Combine the vertical and horizontal filtered images from Sobel filtering. 

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Load the image in grayscale
img = cv.imread('images.jpg', cv.IMREAD_GRAYSCALE)


sobel_x = cv.Sobel(img, cv.CV_64F, 1, 0, 3)
sobel_y = cv.Sobel(img, cv.CV_64F, 0, 1, 3)

sobel_comb = cv.magnitude(sobel_x, sobel_y)
sobel_x = cv.convertScaleAbs(sobel_x)
sobel_y = cv.convertScaleAbs(sobel_y)

comb_result = cv.convertScaleAbs(sobel_comb)


plt.figure(figsize=(15, 15))

plt.subplot(1, 4, 1)
plt.imshow(img, cmap='gray')
plt.title('Original')
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(sobel_x, cmap='gray')
plt.title('Vertical Sobel Filter')
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(sobel_y, cmap='gray')
plt.title('Horizontal Sobel Filter')
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(comb_result, cmap='gray')
plt.title('Combined Sobal Filter')
plt.axis('off')

plt.show()
```

### 1. Should you use bitwise_or function for combining the filtered images? Give proper explanation for your answer.

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = cv.imread('images.jpg', cv.IMREAD_GRAYSCALE)

sobel_x = cv.Sobel(img, -1, 1, 0, 3)
sobel_y = cv.Sobel(img, -1, 0, 1, 3)


sobel_x = cv.convertScaleAbs(sobel_x)
sobel_y = cv.convertScaleAbs(sobel_y)

# bitwise_or
combine_sobel = cv.bitwise_or(sobel_x, sobel_y)

plt.figure(figsize=(15, 15))

plt.subplot(1, 3, 1)
plt.imshow(sobel_x, cmap='gray')
plt.title('Vertical Sobel Filter')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(sobel_y, cmap='gray')
plt.title('Horizontal Sobel Filter')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(combine_sobel, cmap='gray')
plt.title('Combined Sobel using bitwise_or')
plt.axis('off')

plt.show()
```
