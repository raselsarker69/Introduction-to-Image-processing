Bonus-1: Contribution of image thresholding in computer vision. (Notebook/doc is fine enough).
                                                                 
Image thresholding is an important and fundamental technique in computer vision, which is highly effective for image processing, segmentation, object detection, and analysis. It allows objects to be separated from the background, making them easier to analyze and process. Image thresholding is mainly used in tasks such as image analysis, segmentation, and detection.

Uses of Image Thresholding in Computer Vision:

a) Optical Character Recognition (OCR): Image thresholding is used to separate the text from the background in scanned documents, making the text editable.
    
b) Quality Control: In industrial quality control, thresholding is used to check the quality of produced goods, such as identifying defects or imperfections on products.
    
c) Object Segmentation: In various applications, such as autonomous vehicle systems, thresholding is used for object segmentation, allowing vehicles to easily detect obstacles or objects on the road.

Why Image Thresholding is Used in Computer Vision:

a) Image Segmentation: In computer vision, image segmentation is crucial. It helps separate different parts of an image, such as distinguishing a specific object or region from the background.

b) Object Detection: Image thresholding techniques help identify specific objects or parts of an image. It is primarily used for object recognition and tracking, such as detecting license plates, QR codes, or fingerprints.

c) Image Analysis: In many fields, image analysis relies on thresholding, especially in medical imaging, where it is used for accurate identification of organs or tissues.

d) Noise Filtering and Edge Detection: Thresholding helps in noise removal and edge detection in images, which makes the shape and structure of objects more precise and easier to analyze.
    
Image thresholding is a powerful and widely used technique in computer vision, which plays a significant role in image processing, segmentation, and object detection. Its application is highly effective and essential in various industries, medical fields, and automated systems.
