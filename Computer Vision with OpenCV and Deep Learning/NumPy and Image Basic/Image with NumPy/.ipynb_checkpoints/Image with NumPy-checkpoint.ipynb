```python
import cv2 as cv 
import numpy as np 
from PIL import Image
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
img = Image.open('sunflower.jpg')
```

```python
img_arr = np.asarray(img)
```

```python
type(img_arr)
```

```python
img_arr.shape   # 1000x800=(HeightxWidth) pixel and 3 color channels that means RGB
```

```python
plt.imshow(img_arr)
```

```python
img_arr[:, :, 0]
```

```python
plt.imshow(img_arr[:, :, 0], cmap='gray')
```

```python
plt.imshow(img_arr[:, :, 1], cmap='gray')
```

```python
plt.imshow(img_arr[:, :, 2], cmap='gray')
```

```python
plt.imshow(img_arr[:, :, 0], cmap='Reds')    # 0 Red channels,  [:, :, 0] means all row, all coloum, and channels
plt.imshow(img_arr[:, :, 1], cmap='Greens')     
plt.imshow(img_arr[:, :, 2], cmap='Blues')    

# Plotting the three channels separately
plt.figure(figsize=(12, 4))

# Red Channel
plt.subplot(1, 3, 1)
plt.imshow(img_arr[:, :, 0], cmap='Reds')
plt.title('Red Channel')
plt.axis('off')

# Green Channel
plt.subplot(1, 3, 2)
plt.imshow(img_arr[:, :, 1], cmap='Greens')
plt.title('Green Channel')
plt.axis('off')

# Blue Channel
plt.subplot(1, 3, 3)
plt.imshow(img_arr[:, :, 2], cmap='Blues')
plt.title('Blue Channel')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
img_arr = img_arr.copy()  # Create a writable copy of the image array
```
