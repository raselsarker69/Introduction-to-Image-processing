```python
import cv2 as cv 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
img = cv.imread('sunflower1.jpg')
```

```python
plt.imshow(img)
```

```python
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2HSV)
plt.imshow(img_rgb)
```

```python
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2HLS)
plt.imshow(img_rgb)
```

```python
img_rgb = cv.cvtColor(img, cv.COLOR_RGB2HLS)
plt.imshow(img_rgb)
```
