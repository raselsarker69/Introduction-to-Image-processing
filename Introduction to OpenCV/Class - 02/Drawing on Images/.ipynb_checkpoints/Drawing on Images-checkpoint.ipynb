```python
import cv2 as cv 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
blank_img = np.zeros(shape=(512,512,3),dtype=np.int16)
```

```python
plt.imshow(blank_img)
```

```python
blank_img.shape
```

```python
print(blank_img)
```

# Explaination:
- blank_img: The image (or canvas) where the rectangle will be drawn.
- (384, 0): The top-left corner of the rectangle (x=384, y=0).
- (510, 150): The bottom-right corner of the rectangle (x=510, y=150).
- color=(0, 255, 0): The color of the rectangle, which is green in BGR format (Blue, Green, Red).
- thickness=10: The thickness of the rectangle's border in pixels.

```python
cv.rectangle(blank_img, (384,10), (510,150), color=(0,255,0), thickness=10)
```

```python
plt.imshow(blank_img)
```

```python
blank_img.shape
```

```python
cv.rectangle(blank_img, (200,200), (300,300), color=(0,0,255), thickness=10)
```

```python
plt.imshow(blank_img)
```

```python
cv.circle(blank_img, center=(100,100), radius=50, color=(255,0,0), thickness=8)
plt.imshow(blank_img)
```

```python
cv.circle(blank_img, center=(400,400), radius=50, color=(255,0,0), thickness=-1)
plt.imshow(blank_img)
```

```python
cv.line(blank_img, (0,0), (512,512), color=(231,145,255), thickness=10)
plt.imshow(blank_img)
```

```python
font = cv.FONT_HERSHEY_SIMPLEX
cv.putText(blank_img, text='sarker', org=(10,500), fontFace=font, fontScale=4, color=(123,231,255), thickness=5, lineType=cv.LINE_AA)
plt.imshow(blank_img)
```
