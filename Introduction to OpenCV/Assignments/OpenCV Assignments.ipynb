```python
import cv2 as cv 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
img = cv.imread('sunflower1.jpg')
```

```python
plt.imshow(img)
```

```python
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
```

```python
plt.imshow(img_rgb)
```

```python
new_img = img_rgb.copy()
```

```python
new_img = cv.flip(new_img,0)
plt.imshow(new_img)
```

```python
new_img = cv.flip(new_img,-1)
plt.imshow(new_img)
```

```python
pt1 = (150, 250)
pt2 = (500, 600)
```

```python
cv.rectangle(img_rgb, pt1=pt1, pt2=pt2, color=(0,0,255), thickness=8)
plt.imshow(img_rgb)
```
